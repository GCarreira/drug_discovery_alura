```python
### Usar a função crosstab para criar tabela de frequência.
pd.crosstab(df.dose, df.tempo)
```

```python
## Assim criamos 2 índices. Temos 'dose e tempo' dentro de uma classe maior 'tratamento'.
pd.crosstab([df.dose, df.tempo], df.tratamento)
```

```python
### Neste caso todos os dados da tabela juntos somam 100%. ## Para analisar melhor, precisamos da porcentagem por linhas. Portanto (normalize='index').
pd.crosstab([df.dose, df.tempo], df.tratamento, normalize=True) 
```

```python
### Agora temos os 100% das amostras (com e sem droga) divididas na mesma linha.
pd.crosstab([df.dose, df.tempo], df.tratamento, normalize='index')
```

```python
### Compara a média de resposta do "g0" em relação às doses e o tempo entre aplicações.
pd.crosstab([df.dose, df.tempo], df.tratamento, values= df.g0, aggfunc='mean')
```

```python
corr = df.loc[:, 'g0': 'g50'].corr()
```

```python
import numpy as np 

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
dr = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
dr.head()
```

```python
dr.acat_inhibitor.unique()
```

```python
### Seleciona somente as colunas por tipo, no caso, 'int64' pois só queremos os números, excluindo a coluna "id" que é tipo 'object'.
contagem_moa = dr.select_dtypes('int64').sum().sort_values(ascending = False)
contagem_moa
```

```python
### Outra maneira de selecionar é excluindo a coluna 'id com a função 'drop'.
contagem_moa = dr.drop('id', axis=1).sum().sort_values(ascending = False)
contagem_moa
```

```python
dr.drop('id', axis= 1).sum(axis=1)
```

```python
dr.drop('id', axis= 1).sum(axis = 1).sort_values(ascending=False)
```

```python
dr['n_moa'] = dr.drop('id', axis=1).sum(axis=1)
dr.head()
```

```python
dr.n_moa != 0
```

```python
dr['ativo_moa'] = dr.n_moa != 0
dr
```

```python
### pd.merge(df, dr [['id', 'n_moa', 'ativo_moa']], on = 'id' ) salvar numa variável:

d_comb = pd.merge(df, dr [['id', 'n_moa', 'ativo_moa']], on = 'id' )
d_comb
```

```python
### Consultar se houve algum "ativo" nas amostras com_controle. 
d_comb.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
##O correto é não haver, pois não foi usada qualquer droga nesse grupo.
```

```python
### Consultar quantas amostras ativaram no tratamento com droga.
d_comb.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

```python
comp_princ = d_comb.composto.value_counts().index[:5]
sns.boxplot(data = d_comb.query('composto in @comp_princ'), y= 'g8', x= 'composto', )
```
